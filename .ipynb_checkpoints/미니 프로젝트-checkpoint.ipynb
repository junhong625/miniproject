```python
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/Jun_Hong_Ahn/miniproject/accident.csv', sep=',') # accident.csv파일을 불러오는 코드
data = data.drop(['Unnamed: 0'], axis=1) # Unnamed: 0라는 데이터와 상관없는 열이 있어서 삭제
data.tail(5) 
```

```python
data = data.dropna() # 데이터에서 NAN으로 표시된 데이터가 없는 행들 삭제
data = data.reset_index() # 없어진 행의 index번호가 비어있기 때문에 인덱스를 순서대로 재설정
data = data.drop(['index'],axis=1) # reset_index를 실행하면서 기존의 index 가 컬럼으로 넘어오게 되므로 index 컬럼을 삭제
```

```python
data = data.astype({'사고년도': 'int'}) # data의 '사고년도' 열의 타입을 int로 변경
data = data.astype({'사고년도': 'str'}) # data의 '사고년도' 열의 타입을 int로 변경
data = data.astype({'사고월': 'str'})
data = data.astype({'사고일': 'str'})
```

```python
data
```

```python
year = [] 
for i in range(len(data)):
    year.append(data['사고년도'][i]) # 사고년도를 데이터에서 뽑아와 리스트로 저장
```

```python
month = []
for i in range(len(data)):
    month.append(data['사고월'][i][0:2]) # 사고월을 데이터에서 뽑아와 리스트로 저장
```

```python
day = []
for i in range(len(data)):
    day.append(data['사고일'][i][0:2]) # 사고일을 데이터에서 뽑아와 리스트로 저장
```

```python
date = []
for i in range(len(data)):
    date.append(str(year[i])+'-'+str(month[i])+'-'+str(day[i])) # 사고년도, 사고월, 사고일을 합쳐서 date에 리스트로 저장
```

```python
date[5480:]
```

```python
for i in range(len(data)):
    data['사고년도'][i] = date[i] # data의 사고년도를 date로 변경하여 0000-00-00 형식으로 출력되게 변경
```

```python
data = data.drop(['사고월'], axis=1) # 사고년도에서 모든 날짜 데이터가 출력되므로 사고월과 사고일 열을 삭제
data = data.drop(['사고일'], axis=1)

data
```

## 사고건수, 사망자수, 부상자수 별로 데이터 모아서 분류

```python
A = [] ## 사고건수 리스트
for i in range(len(data)):
    if data['구분'][i] == '사고건수':
        A.append(data['구분값'][i])
        
B = [] ## 사망자수 리스트
for i in range(len(data)):
    if data['구분'][i] == '사망자수':
        B.append(data['구분값'][i])
        
C = [] ## 부상자수 리스트
for i in range(len(data)):
    if data['구분'][i] == '부상자수':
        C.append(data['구분값'][i])
        
data = data.drop_duplicates(['날짜']) ## 중복값 삭제

data.reset_index()
data = data.drop(['index'],axis=1)
data = data.drop(['구분'],axis=1)
data = data.drop(['구분값'],axis=1)

data['사고건수'] = A
data['사망자수'] = B
data['부상자수'] = C

data = data.set_index(['날짜'])
```

## 강수량 데이터

```python
## 강수량 전체데이터 전처리

import csv
import pandas as pd

df = pd.read_csv('extremum_20210915181043.csv', encoding='euc-kr', header=7) 
# 그냥 불러왔을 때 CSV파일에 쓸데 없는 데이터 때문에 에러가 나니 header=7을 통해 csv파일의 7번째 줄부터 데이터를 가져온다.

df = df.drop(['지점명'], axis=1)
df = df.drop(['\t지점번호'], axis=1)
df = df.drop(['1시간최다강수량(mm)'], axis=1)
df = df.drop(['1시간최다강수량시각'], axis=1)
df = df.drop(['Unnamed: 6'], axis=1)
# 내가 원하는 데이터와 상관없는 데이터들은 삭제하여 보기 좋게 데이터를 고친다. 

#df = df.fillna(0)
df = df.dropna()
# df파일에 dropna 메소드를 통해 NA 즉, 데이터가 들어있지 않은 부분의 행을 삭제한다.

df.reset_index(inplace=True)
df = df.drop(['index'], axis=1) 
# reset_index 메소드를 실행하면서 생겨난 index column을 삭제한다.

df.rename(columns = {'일시':'날짜'}, inplace=True)
df = df.set_index(['날짜'])
# 보기 좋게 index를 날짜로 설정한다
```

```python
H = df['강수량(mm)'] > 80 # 강수량이 80mm 이상인 데이터를 H로 저장 
df[H]
```

```python
M = df['강수량(mm)'] < 5 # 강수량이 5mm 이하인 데이터를 M으로 저장
df[M]
```

```python
Z = df['강수량(mm)'] == 0.0 # 강수량이 0mm인 데이터
df[Z]
```

## 교통사고

```python
accident = pd.read_csv('data/2020.txt', sep='\t')
for a in range(3, 39, 3):
    accident['사고월'][a] = '2020년 ' + accident['사고월'][a]
accident.transpose()
for a in range(3):
    accident = accident.drop(a, axis=0)
accident.transpose()
accident.reset_index(inplace=True)
accident = accident.drop(['index'], axis=1)
E = accident
```

```python
df[H].index
```

```python
accident.index
```

```python
z = []
for i in df[Z].index :
    for j in accident.index :
        if i == j:
            z.append(accident.loc[j])
            
m = []
for i in df[M].index :
    for j in accident.index :
        if i == j:
            m.append(accident.loc[j])
        
h = []
for i in df[H].index :
    for j in accident.index :
        if i == j:
            h.append(accident.loc[j])
```

```python
accident_total = 0
for i in range(len(h)):
    accident_total += h[i]['사고건수']
print(accident_total)
print(accident_total/len(h)) # 강수량이 높은 날들의 평균 사고건수
```

```python
death_total = 0
for i in range(len(h)):
    death_total += h[i]['사망자수']
print(death_total)
print(death_total / len(h)) # 평균 사망자수
```

```python
injury_total = 0
for i in range(len(h)):
    injury_total += h[i]['부상자수']
print(injury_total)
print(injury_total / len(h)) # 평균 부상자수
```

```python
zero_accident_total = 0
for i in range(len(z)):
    zero_accident_total += z[i]['사고건수']
print(zero_accident_total)
print(zero_accident_total/len(z))
```

```python
zero_death_total = 0
for i in range(len(z)):
    zero_death_total += z[i]['사망자수']
print(zero_death_total)
print(zero_death_total / len(z)) # 평균 사망자수
```

```python
zero_injury_total = 0
for i in range(len(z)):
    zero_injury_total += z[i]['부상자수']
print(zero_injury_total)
print(zero_injury_total / len(z)) # 평균 부상자수
```

```python
Middle_accident_total = 0
for i in range(len(m)):
    Middle_accident_total += m[i]['사고건수']
print(Middle_accident_total)
print(Middle_accident_total/len(m))
```

```python
len(m)
```

```python
Middle_death_total = 0
for i in range(len(m)):
    Middle_death_total += m[i]['사망자수']
print(Middle_death_total)
print(Middle_death_total / len(m)) # 평균 사망자수
```

```python
Middle_injury_total = 0
for i in range(len(m)):
    Middle_injury_total += m[i]['부상자수']
print(Middle_injury_total)
print(Middle_injury_total / len(m)) # 평균 부상자수
```

```python
zero = [zero_accident_total/len(z), zero_death_total/len(z), zero_injury_total/len(z)]
middle = [Middle_accident_total/len(m), Middle_death_total/len(m), Middle_injury_total/len(m)]
high = [accident_total/len(h), death_total/len(h), injury_total/len(h)]
```

```python
middle
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import platform 

print(platform.system()) # 플랫폼 확인 # Window #한글표시
if platform.system() == 'Windows': 
    matplotlib.rc('font', family='Malgun Gothic') 
elif platform.system() == 'Darwin': # Mac 
    matplotlib.rc('font', family='AppleGothic') 
else: #linux 
    matplotlib.rc('font', family='NanumGothic') 

# 그래프에 마이너스 표시가 되도록 변경 
matplotlib.rcParams['axes.unicode_minus'] = False

r = ['사고건수', '사망자수', '부상자수']
x = np.arange(len(zero))
plt.figure(figsize=(13,13))
plt.bar(x-0.2, zero, label='강수량 = 0', width=0.2, color='gray')
plt.bar(x-0.0, middle, label='강수량 < 5', width=0.2, color='silver')
plt.bar(x+0.2, high, label='강수량 > 80', width=0.2, color='black')
plt.title('강수량에 따른 평균 교통사고수')
plt.xticks(x, r)
plt.ylabel('(평균 사고건수), (사망자,부상자 수)', fontsize=15)
plt.legend()
```
